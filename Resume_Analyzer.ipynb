```python
def create_prompt(resume_text):

    prompt = f"""
You are an ATS Resume Analyzer.

Analyze the following resume.

Resume:
{resume_text}

Give the output in this format:

1. ATS Score (Out of 100)

2. Strengths

3. Weaknesses

4. Missing Skills

5. Improvement Suggestions

6. Suitable Job Roles

7. Career Advice
"""

    return prompt
```

```python
def analyze_resume(resume_text):

    prompt = create_prompt(resume_text)

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
                "role":"user",
                "content":prompt
            }
        ],
        temperature=0.3
    )

    return response.choices[0].message.content
```

```python
import fitz
def extract_text(pdf_file):
    document = fitz.open(pdf_file)
    text = ""

    for page in document:
        text += page.get_text()
    document.close()

    return text
```

```python
def process_resume(pdf_file):
    resume_text = load_resume(pdf_file)
    if resume_text is None:
        return "Not able to read"
    report = analyze_resume(resume_text)
    return report
```

```python
def process_resume(pdf_file):
    try:
        resume_text = load_resume(pdf_file)
        if resume_text is None:
            return "Not able to read"
        report = analyze_resume(resume_text)
        return report
    except Exception as e:
        return str (e)
```

```python
def resume_statistics(text):
    stats = f"""
    Characters : {len(text)}

    WORDS : {len(text.split())}

    Lines :  {len(text.splitlines())}
    
    """
    return stats
```

```python
print(resume_statistics(resume))
```

```python
def generate_complete_report (pdf_file):
    resume_text = load_resume(pdf_file)
    if resume_text is None:
        return "Not able to read"
    
    stats = resume_statistics(resume_text)

    report = analyze_resume(resume_text)

    return stats + "\n\n\n" + report
    
```

```python
print(generate_complete_report("Aryan_Singh_Jadaun.pdf"))
```

```python
def analyze_upload_resume(pdf_file):
    if pdf_file is None :
        return "Please Upload a resume"

    try:
        report = generate_complete_report(pdf_file)
        return report
    except Exception as e:
        return f"Error : {e}"
```

```python
def generate_linkedin_post(pdf):
    resume = extract_text(pdf.name)

    prompt = f"""
Create a professional LinkedIn post based on this resume.

{resume}
"""

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ]
    )

    linkedin_post = response.choices[0].message.content

    with open("linkedin_post.txt", "w", encoding="utf-8") as file:
        file.write(linkedin_post)

    return linkedin_post
```

```python
import gradio as gr

with gr.Blocks() as app:

    gr.Markdown("# 🤖 AI Resume Analyzer")

    # ---------- Resume Analyzer ----------
    with gr.Tab("Resume Analyzer"):

        resume = gr.File(label="Upload Resume", file_types=[".pdf"])

        output1 = gr.Textbox(label="Analysis", lines=15)

        gr.Button("Analyze").click(
            fn=analyze_upload_resume,
            inputs=resume,
            outputs=output1
        )

    # ---------- LinkedIn Post ----------
    with gr.Tab("LinkedIn Post"):

        resume2 = gr.File(label="Upload Resume", file_types=[".pdf"])

        output2 = gr.Textbox(label="LinkedIn Post", lines=10)

        gr.Button("Generate").click(
            fn=generate_linkedin_post,
            inputs=resume2,
            outputs=output2
        )

    # ---------- Merge PDF ----------
    with gr.Tab("Merge PDF"):

        pdf1 = gr.File(label="PDF 1", file_types=[".pdf"])

        pdf2 = gr.File(label="PDF 2", file_types=[".pdf"])

        output3 = gr.File(label="Merged PDF")

        gr.Button("Merge").click(
            fn=merge_pdf,
            inputs=[pdf1, pdf2],
            outputs=output3
        )

app.launch()
```

```python
def analyze_upload_resume(pdf_file):

    if pdf_file is None:
        return "Please Upload a Resume"

    try:
        report = generate_complete_report(pdf_file)

        with open("history.txt", "a", encoding="utf-8") as file:
            file.write("\n")
            file.write("="*60)
            file.write("\n")
            file.write(report)
            file.write("\n")

        return report

    except Exception as e:
        return f"Error : {e}"
```

```python
def merge_pdf(pdf1, pdf2):

    merger = PdfMerger()

    merger.append(pdf1.name)
    merger.append(pdf2.name)

    output = "Merged_Notes.pdf"

    merger.write(output)

    merger.close()

    return output
```
